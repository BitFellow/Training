```python
#Nested If-else statement
score = int(input("Enter your score: "))
if score >= 0 and score <= 100:
    if score >= 90:
        print("Distinction")
    elif score >= 80:
        print("A Grade")
    elif score >= 70:
        print("B Grade")
    elif score >= 60:
        print("C Grade")
    elif score > 50:
        print("D Grade")
    elif score == 50:
        print("E Grade")
    else:
        print("U Grade")
else:
    print("Enter a valid score between 0 and 100")
```

```python
products = ["MacBook", "iPad"]

colors = ["Silver", "Space Gray", "Gold"]

for product in products:
    for color in colors:
        print(product, "-", color)
```

```python
climate_data = [
    {"city": "City A", "temperature": 25, "carbon_footprint": 500},
    {"city": "City B", "temperature": 30, "carbon_footprint": 350},
    {"city": "City C", "temperature": 22, "carbon_footprint": 600},
    {"city": "City D", "temperature": 15, "carbon_footprint": 200},
    {"city": "City E", "temperature": 28, "carbon_footprint": 450},
]

threshold = 300

print("Cities with carbon footprint above 300 kg:\n")

for data in climate_data:
    if data["carbon_footprint"] > threshold:
        print(f"{data['city']} -> {data['carbon_footprint']} kg")
```
